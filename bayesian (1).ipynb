```python
# Selecting relevant predictors and the outcome variable for Basel
basel_data = weather_data[['BASEL_temp_mean', 'BASEL_precipitation', 'BASEL_humidity']].copy()
basel_bbq = bbq_labels['BASEL_BBQ_weather'].astype(int)  # Converting boolean to integer for logistic regression

# Centering the data by subtracting the mean
basel_data_centered = basel_data.apply(lambda x: x - x.mean())

# Combining centered predictors with the outcome
basel_data_centered['BBQ'] = basel_bbq

# Display the first few rows to check the centered data
basel_data_centered.head()
```

```python
import pymc as pm  # Note that PyMC4 uses 'pymc' instead of 'pymc3'

# Model setup remains similar, but with updated library calls
with pm.Model() as model:
    # Define priors for the intercept and slopes
    alpha = pm.Normal('alpha', mu=0, sigma=10)
    beta_temp = pm.Normal('beta_temp', mu=0, sigma=10)
    beta_precip = pm.Normal('beta_precip', mu=0, sigma=10)
    beta_humid = pm.Normal('beta_humid', mu=0, sigma=10)

    # Linear model combining predictors and intercept
    mu = alpha + beta_temp * basel_data_centered['BASEL_temp_mean'] + \
         beta_precip * basel_data_centered['BASEL_precipitation'] + \
         beta_humid * basel_data_centered['BASEL_humidity']

    # Define the Bernoulli likelihood
    obs = pm.Bernoulli('obs', logit_p=mu, observed=basel_data_centered['BBQ'])

    # Sampling from the posterior distribution using the default sampler
    trace = pm.sample(5000)

# This line is optional: It helps to print the summary of the trace
pm.summary(trace).round(4)
```

```python
import arviz as az

# Create trace plots for each parameter
az.plot_trace(trace)
plt.show()

# Print the summary statistics of the posterior distribution
summary_stats = pm.summary(trace)
print(summary_stats)

# Generate posterior predictive samples
with model:
    posterior_predictive = pm.sample_posterior_predictive(trace, var_names=['obs'])

# Compare observed and predicted outcomes
_, ax = plt.subplots(figsize=(12, 6))
az.plot_ppc(az.from_pymc3(posterior_predictive=posterior_predictive, model=model))
plt.show()
```

```python
# Calculate probabilities over a grid of temperatures (as an example)
import numpy as np
temp_range = np.linspace(basel_data_centered['BASEL_temp_mean'].min(), basel_data_centered['BASEL_temp_mean'].max(), 100)
probabilities = 1 / (1 + np.exp(-(summary_stats.at['alpha', 'mean'] +
                                   summary_stats.at['beta_temp', 'mean'] * temp_range)))

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(temp_range, probabilities, label='Probability of BBQ-Suitable Day')
plt.scatter(basel_data_centered['BASEL_temp_mean'], basel_data_centered['BBQ'], alpha=0.2)
plt.title('Effect of Temperature on BBQ Suitability')
plt.xlabel('Temperature (centered)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from scipy.special import expit as logistic_sigmoid
import matplotlib.pyplot as plt

# Load the datasets
bbq_labels = pd.read_csv('/content/weather_prediction_bbq_labels.csv')
weather_data = pd.read_csv('/content/weather_prediction_dataset.csv')

# Merge the data on the 'DATE' column
data = pd.merge(weather_data, bbq_labels, on='DATE')

# Define cities and predictors
cities = ['BASEL', 'BUDAPEST', 'DRESDEN', 'DUSSELDORF']
predictors = ['_temp_mean', '_precipitation', '_humidity']

# Prepare to store models and predictions
models = {}
curves = {pred: [] for pred in predictors}

# Fit logistic regression models for each city and predictor
for city in cities:
    for pred in predictors:
        feature = city + pred
        if feature in data.columns:
            X = data[[feature]].dropna()
            y = data[f'{city}_BBQ_weather'].dropna()
            if len(X) == len(y):
                model = LogisticRegression()
                model.fit(X, y)
                models[(city, pred)] = model
                # Generate sigmoid curve data
                X_range = np.linspace(X.min(), X.max(), 300)
                y_prob = logistic_sigmoid(model.intercept_ + model.coef_ * X_range)
                curves[pred].append((X_range, y_prob))

# Now let's plot these curves
for pred in predictors:
    plt.figure(figsize=(8, 6))
    for X_range, y_prob in curves[pred]:
        plt.plot(X_range, y_prob, color='grey', alpha=0.5)
    # Calculate and plot the average curve
    all_y_probs = np.array([y_prob for _, y_prob in curves[pred]])
    mean_prob = np.mean(all_y_probs, axis=0)
    plt.plot(X_range, mean_prob, color='black', linewidth=2)
    plt.title(f'Effect of {pred[1:]} on BBQ-Suitable Weather Across Cities')
    plt.xlabel(f'{pred[1:]}')
    plt.ylabel('Probability of BBQ-Suitable Weather')
    plt.show()
```
